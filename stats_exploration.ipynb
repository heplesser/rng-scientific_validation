```python
from plot_statistics import Validate
import matplotlib.pyplot as plt
import statsmodels.api as sm
import scipy
```

```python
v = Validate(analysis_path='./data/mam', quantity='rates', 
             model='mam', versions=['master-rng', '2.20.1'])
v.load_data()
r2 = v.data['2.20.1'].loc['V1', '4', 'E'].values
r3 = v.data['master-rng'].loc['V1', '4', 'E'].values
plt.plot(r2[:28], r3, 'o');
```

```python
sm.qqplot(r2, label='2.20.1')
sm.qqplot(r3, label='rng')
plt.legend();
```

```python
sm.qqplot_2samples(r2, r3, line='45', xlabel='Rates, NEST 3 RNG', ylabel='Rates, NEST 2 RNG');
```

```python
scipy.stats.ks_2samp(r2, r3)
```

```python
scipy.stats.epps_singleton_2samp(r2, r3)
```

```python
scipy.stats.anderson_ksamp([r2, r3])
```
